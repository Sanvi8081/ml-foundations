```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=3)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.plot(x,y2,color='b',linestyle='-',linewidth=3)
plt.title("line plot with 2 lines")
plt.xlabel("x-axis")
plt.ylabel("y-axiz")

plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,y1,color="orange",linestyle='-',linewidth=2)
plt.show()
```

```python
student={"Soha":89,"Palak":79,"Dhruv":82}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("BAR PLOT")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color='g')
plt.title("BAR PLOT")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
plt.scatter(x,y,marker="^",c="g",s=200)
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
z=[9,8,7,6,4]
plt.scatter(x,y,marker="^",c="g",s=100)
plt.scatter(x,z,marker="*",c="b",s=100)
plt.show()
```

```python
x=[10,20,30,40,50]
y=[1,2,3,4,5]
z=[9,8,7,6,4]
plt.subplot(1,2,1)
plt.scatter(x,y,marker="^",c="g",s=100)
plt.subplot(1,2,2)
plt.scatter(x,z,marker="*",c="b",s=100)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[12,14,16,13,11]
z=[6,7,8,9,5]
data=list([x,y,z])
plt.boxplot(data)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[12,14,16,13,11]
z=[6,7,8,9,5]
data=list([x,y,z])
plt.violinplot(data,showmedians=True)
plt.show()
```
