```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#downloads stock data
stock = "AAPL"  # Change stock here
data = yf.download(stock, start="2020-01-01", end="2024-01-01")
data.head()
```

```python
#OHLC data analysis
print("First 5 rows:")
print(data.head())

print("\nData Info:")
print(data.info())

print("\nStatistical Summary:")
print(data.describe())
```

```python
#moving average
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()

data[['Close', 'MA_5', 'MA_10']].head()
```

```python
#daily returns
data['Returns'] = data['Close'].pct_change()
data[['Close', 'Returns']].head()
```

```python
#volitality
data['Volatility'] = data['Returns'].rolling(window=5).std()
data[['Returns', 'Volatility']].head()
```

```python
#high-low open-close
data['High-Low'] = data['High'] - data['Low']
data['Open-Close'] = data['Open'] - data['Close']

data.head()
```

```python
#saving
data.to_csv("stock_data.csv")
print("File saved as stock_data.csv")
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['MA_5'], label='5 Day MA')
plt.plot(data['MA_10'], label='10 Day MA')
plt.title("Stock Price and Moving Average")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Returns'], label='Daily Returns')
plt.title("Daily Returns")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Volatility'], label='Volatility')
plt.title("Volatility")
plt.legend()
plt.show()
```
